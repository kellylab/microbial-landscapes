```python
# Load dataset
import pandas as pd
df = pd.read_csv('BIGRAPA_COGS.csv', sep=',',header=0, index_col=0)
```

```python
# Define a similarity metric between different conditions
```

```python
x = df.loc['COG999'].as_matrix().squeeze()
y = df.loc['COG998'].as_matrix().squeeze()
np.dot(x,y) / np.sqrt( np.dot(x,x) * np.dot(y,y) )
```

```python
def similarity(x, y):
    return np.dot(x,y) / np.sqrt( np.dot(x,x) * np.dot(y,y) )
```

```python
data = df.as_matrix()
[nrows, ncols] = data.shape
S = np.zeros([nrows, nrows], np.float32)
for i in range(nrows):
    for j in range(i, nrows):
        S[i,j] = S[j,i] = similarity(data[i,:], data[j,:])
```

```python
np.save('similarity.npz', S)
```
