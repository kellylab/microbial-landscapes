```python
# Define a similarity metric between different conditions
def similarity(x, y):
    numer = np.dot(x,y)
    x2 = np.dot(x,x)
    y2 = np.dot(y,y)
    denom = np.sqrt(x2 * y2)
    if (x2 == 0.0) and (y2 == 0.0): 
        s = 1.0
    elif (denom == 0.0):
        s = 0.0
    else:
        s = numer / denom
    return s

# Compute the similarity matrix on a subset of the data
#S = np.zeros([nrows, nrows], np.float32)
#for i in range(nrows):
#    for j in range(i, nrows):
#        S[i,j] = S[j,i] = similarity(data[i,:], data[j,:])
```

```python
# Spectral clustering
import msmbuilder.cluster
```

```python
# transform to log data
data = df.as_matrix()[:ndata,:] # subset
log_zero = np.log(data[data > 0.0].min() / 2.0)
log_data = np.log(data)
log_data[log_data == -np.inf] = log_zero # for no reads
```

```python
# Plot the original data
fig, axes = plt.subplots(1, 2, figsize=(12,30))  

ax = axes.flatten()[0]
plt.sca(ax)
img = ax.imshow(data, interpolation='nearest')
ax.imshow(data);
ax.set_aspect('auto')    
plt.title('original');
plt.xticks([]);
plt.yticks([]);

ax = axes.flatten()[1]
plt.sca(ax)
ax.imshow(log_data);
ax.set_aspect('auto')    
plt.title('log reads');
plt.xticks([]);
plt.yticks([]);
```

```python
# Fit using precomputed similarity matrix
#clustering = msmbuilder.cluster.SpectralClustering(n_clusters=25, affinity='precomputed', assign_labels='kmeans')
#clustering.fit([S])

# Alternative: Fit using radial basis function distance
clustering = msmbuilder.cluster.SpectralClustering(n_clusters=10, affinity='rbf', gamma=0.01, assign_labels='kmeans', n_init=500)
clustering.fit([log_data])
```

```python
# Reorder the similarity matrix S and data to group gene sets in the same cluster next to each other
labels = clustering.labels_[0]
order = np.argsort(labels)
S = clustering.affinity_matrix_
S_sorted = S[order,:]
S_sorted = S_sorted[:,order]
data_sorted = data[order,:]
log_data_sorted = log_data[order,:]
labels_sorted = labels[order]
```

```python
# Plot the original and clustered similarity matrices
plt.figure(figsize=[15,30])
plt.subplot(1,2,1);
plt.imshow(S);
plt.title('original similarity matrix');
plt.subplot(1,2,2);
plt.imshow(S_sorted);
plt.title('clustered similarity matrix');
```

```python
# Plot dominant eigenvalues
from scipy.sparse.linalg import eigsh
kmax = 50 # max num of eigenvalues
[eigenvalues, eigenvectors] = eigsh(S, k=kmax)
eigenvalues = eigenvalues[::-1]

plt.subplot(1,2,1)
plt.plot(eigenvalues, '.');
plt.xlabel('eigenvalue number');
plt.ylabel('eigenvalue');
plt.title('leading eigenvalues');

plt.subplot(1,2,2)
plt.plot(eigenvalues, '.');
plt.axis([0, kmax, 0, 10])
plt.xlabel('eigenvalue number');
plt.ylabel('eigenvalue');
plt.title('zoomed');
```

```python
# Plot the original and clustered data matrices (using log reads for clarity)
fig, axes = plt.subplots(1, 3, figsize=(12,30))  

ax = axes.flatten()[0]
plt.sca(ax)
ax.imshow(log_data);
ax.set_aspect('auto')    
plt.title('log reads');
plt.xticks([]);
plt.yticks([]);

ax = axes.flatten()[1]
plt.sca(ax)
im = ax.imshow(log_data_sorted);
ax.set_aspect('auto')    
plt.title('log reads sorted');
plt.xticks([]);
plt.yticks([]);

ax = axes.flatten()[2]
plt.sca(ax)
plt.imshow(np.tile(np.reshape(labels_sorted, [ndata,1]), [1,28]));
ax.set_aspect('auto')    
plt.title('clusters');
plt.xticks([]);
plt.yticks([]);
plt.colorbar(im)

plt.tight_layout()
```

```python
nclusters = labels.max() + 1
for cluster in range(nclusters):
    indices = list(np.where(labels == cluster)[0])
    print('Cluster %d / %d' % (cluster, nclusters))
    gene_names = df.index[indices].tolist()
    print(gene_names)
    print('')
```
